Why Pydantic?
To solve issues which arises because of dynamic typing in Python. Object is created without throwing  and this is covered later on as Python has dynamic typing features unlike C#/Java,
To learn more: https://youtu.be/XIdQ6gO3Anc


```python
# This should fail but because of dynamic typing in Python it will 
# not throw any error and will create the object successfully.

class UserModel_Without_Pydantic:
    name: str
    email: str
    account_id: int

    def __init__(self, name: str, email: str, account_id: int):
        self.name = name
        self.email = email
        self.account_id = account_id

user = UserModel_Without_Pydantic(name="John Doe", email="john.doe@example.com",
                                  account_id="eee")
print(user.account_id)
```

```python
# To solve the above issue, we can use Pydantic which will validate the data types of the 
# fields and throw an validation error if the data types are not correct, so we will 
# come to know about this early before deploying to any environment.

from pydantic import BaseModel as PydanticBaseModel

class UserModel(PydanticBaseModel):
    name: str
    email: str
    account_id: int
    
# Throws Validation errors as account_id is expected to be an integer
user = UserModel(name="John Doe", email="john.doe@example.com",
                                  account_id="eee")

print(user.name)
```

```python
from pydantic import BaseModel as PydanticBaseModel, EmailStr

class UserModel_Email_Validator(PydanticBaseModel):
    name: str
    email: EmailStr
    account_id: int
    
# Throws Validation errors as account_id is expected to be an integer
# To fix change email to a valid email address like john.doe@example.com
user = UserModel_Email_Validator(name="John Doe", email="john.doe",
                                  account_id=12)

print(user.name)
```
